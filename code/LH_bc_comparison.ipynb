```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import array
```

```python
root_path = '/home/gemeinsam_tmp/VACILT/ENSO_data/'
xr.set_options(display_style='html')  
```

```python
nb = 36
lat = np.arange(-87.5,-87.5+5.*nb,5.,float)
lev = np.arange(1.421,160.,2.842,float)
nz = 10
```

# m = 0

```python
phase_ls = ['EL','La']

da_ls = []
for phase in phase_ls:
    fileobj=open(f'{root_path}lh_{phase}_m0.dx',mode='rb') 	
    binvalues = array.array('f')
    binvalues.fromfile(fileobj, 2*nz*nb*2)
    uvt = np.array(binvalues, np.float32)
    uvt = np.reshape(uvt,(2,nz,nb,2))
    fileobj.close()
    
    da_m0 = xr.DataArray(uvt, dims = ['var','lev','lat','tides'])
    da_m0['tides'] = [24,12]
    da_m0['lev'] = lev[:10]
    da_m0['lat'] = lat
    da_m0['var'] = ['amp','pha']
    
    da_ls.append(da_m0)
    
da_m0_all = xr.concat(da_ls, dim = 'phase')
da_m0_all['phase'] = phase_ls
da_m0_all
```

```python
sel_dict = dict(var = 'amp', lat = slice(-40,40), lev = slice(None,15))
da_m0_all.sel(**sel_dict).plot.contourf(col ='tides', row = 'phase', 
                                         x = 'lat', vmin = 0, 
                                         levels = 21, vmax = 0.32)
```

# m = 1

```python
phase_ls = ['EL','La']

da_ls = []
for phase in phase_ls:
    fileobj=open(f'{root_path}lh_{phase}_m1.dx',mode='rb') 	
    binvalues = array.array('f')
    binvalues.fromfile(fileobj, 4*nz*nb*3)
    uvt = np.array(binvalues, np.float32)
    uvt = np.reshape(uvt,(2,2,nz,nb,3))
    fileobj.close()
    
    da_m0 = xr.DataArray(uvt[...,:2], dims = ['var','wave','lev','lat','tides'])
    da_m0['tides'] = [24,12]
    da_m0['lev'] = lev[:10]
    da_m0['lat'] = lat
    da_m0['var'] = ['amp','pha']
    da_m0['wave'] = ['westward','eastward']
    
    da_ls.append(da_m0)
    
da_m0_all = xr.concat(da_ls, dim = 'phase')
da_m0_all['phase'] = phase_ls
da_m0_all
```

## El Nino

```python
sel_dict = dict(var = 'amp', lat = slice(-40,40), lev = slice(None,15), phase = 'EL')
da_m0_all.sel(**sel_dict).plot.contourf(col ='tides', row = 'wave', 
                                         x = 'lat', vmin = 0, vmax = 0.32,
                                         levels = 21)
```

## La Nina

```python
sel_dict = dict(var = 'amp', lat = slice(-40,40), lev = slice(None,15), phase = 'La')
da_m0_all.sel(**sel_dict).plot.contourf(col ='tides', row = 'wave', 
                                         x = 'lat', vmin = 0, vmax = 0.32,
                                         levels = 21)
```
